```python
# make a plot of salary dictonary 

salaries = {"Salary One": 1000,
          "Salary Two": 3000,
          "Salary Three": 5000}


fig, ax = plt.subplots(figsize=(5,5))
ax.bar(salaries.keys(), salaries.values())
ax.set(title="Salary Analysis",
      xlabel="Monthly Salary",
      ylabel="Salary Range");
 
```

###  Which one should be use? (pyplot vs matplotlib oop)

* When ploting something quickly, okay to use the pyplot method
* When ploting something more advance, use the oop method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
# pyplot method

over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");

ax.set_xlim([45, 100]);
```

```python
# oop method

fig, ax = plt.subplots(figsize=(10, 6))

over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)

# ax.set_xlim([45, 100]);
```

```python
over_50.target.values
```

```python
## OOP from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data

scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle="--",
          color="g");
```

```python
# Set Style
plt.style.use("seaborn-v0_8-whitegrid")

# Create plot for chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0

scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap="winter") # change the color schema

#change the x axis limits
ax0.set_xlim([50, 80])

# Customize plot for ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")


# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle="--",
           color="r")



# add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                     cmap="winter")

# Change ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Customize to ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add horizontal line
ax1.axhline(y=over_50['thalach'].mean(), 
            linestyle="--",
           color="r");


# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
 
plt.savefig("images/heart-disease-analysis.png");
```
