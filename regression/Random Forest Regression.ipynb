### Note: Ensemble learning and random forest regression are same


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Drug_Effectiveness.csv")
df
```

```python
# All independent variable
X = df.iloc[:, 0:1].values

# This is just dependent variable
y = df['Effectiveness'].values
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

# n_estimators means number of trees and random state means same result show evety time

reg = RandomForestRegressor(n_estimators=1000, random_state=0)

reg.fit(X, y);
```

```python
y_pred = reg.predict(X)

print(y_pred)
```

```python
print(reg.predict([[ 12 ]]))
```

```python
plt.scatter(X, y, color="red")
plt.plot(X, y_pred, color="yellow")
plt.title("Dosage VS Effectiveness")
plt.xlabel("Dosage (mg)")
plt.ylabel("Effectiveness (%)");
```
