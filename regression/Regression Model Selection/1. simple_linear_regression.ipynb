## Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing The Dataset

```python
# df = pd.read_csv("/content/Drug_Effectiveness.csv");
df = pd.read_csv("/content/Salary_Data.csv")
df
```

```python
# This is independent variable
X = df.iloc[:, :-1].values

# This is dependent variable
y = df.iloc[:, -1].values
```

## Splitting the dataset into the training set and test set

* X_train means independent variable train
* X_test means independent variable test
* y_train means dependent variable train
* y_test means dependent variable test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train);
```

```python
reg.score(X_test, y_test)
```

```python
print(np.concatenate((X_test, y_test.reshape(len(y_test), 1)), 1) )
```

```python
y_pred = reg.predict(X_test)
np.set_printoptions(precision=2)
```

```python
y_pred = y_pred.reshape(len(y_pred), 1)
y_test = y_test.reshape(len(y_test), 1)


print(np.concatenate((y_pred, y_test), 1))
```

```python
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, reg.predict(X_train));
plt.title("Experience Vs Salary")
plt.xlabel("Experience (year)")
plt.ylabel("Salary (TK)")
```

```python
# OOP Plot
fig, ax = plt.subplots(figsize=(6, 5))

ax.scatter(X_train, y_train, color="green");
ax.scatter(X_train, reg.predict(X_train), color="red")

# Customize the plot
ax.set(title="Experience VS Salary",
       xlabel="Expericence (Year)",
       ylabel="Salary (TK)");


plt.savefig("images/salary.png");
```

## Evaluating the model Performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
